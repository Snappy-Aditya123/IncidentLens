```python
import dpkt
import socket
import csv

pcap_file = r"C:\Users\adity\Desktop\AI\light\competetions\elasticsearch\data\SSDP_Flood_pcap.pcap\SSDP_Flood_pcap.pcap"
output_file = "ssdp_packets_rich.csv"

def inet_to_str(inet):
    try:
        return socket.inet_ntop(socket.AF_INET, inet)
    except:
        return None

prev_ts = None

with open(pcap_file, 'rb') as f, open(output_file, 'w', newline='') as out:
    pcap = dpkt.pcapng.Reader(f)
    writer = csv.writer(out)

    writer.writerow([
        "packet_index",
        "timestamp",
        "inter_arrival_time",
        "src_ip",
        "dst_ip",
        "src_port",
        "dst_port",
        "protocol",
        "ttl",
        "ip_header_len",
        "tcp_flags",
        "udp_length",
        "payload_length",
        "packet_length"
    ])

    for i, (ts, buf) in enumerate(pcap):
        try:
            eth = dpkt.ethernet.Ethernet(buf)
        except:
            continue

        inter_arrival = 0
        if prev_ts is not None:
            inter_arrival = ts - prev_ts
        prev_ts = ts

        src_ip = None
        dst_ip = None
        src_port = None
        dst_port = None
        protocol = None
        ttl = None
        ip_header_len = None
        tcp_flags = None
        udp_length = None
        payload_length = None

        if isinstance(eth.data, dpkt.ip.IP):
            ip = eth.data
            src_ip = inet_to_str(ip.src)
            dst_ip = inet_to_str(ip.dst)
            protocol = ip.p
            ttl = ip.ttl
            ip_header_len = ip.hl * 4

            payload_length = len(ip.data)

            if isinstance(ip.data, dpkt.tcp.TCP):
                src_port = ip.data.sport
                dst_port = ip.data.dport
                tcp_flags = ip.data.flags
                payload_length = len(ip.data.data)

            elif isinstance(ip.data, dpkt.udp.UDP):
                src_port = ip.data.sport
                dst_port = ip.data.dport
                udp_length = ip.data.ulen
                payload_length = len(ip.data.data)

        writer.writerow([
            i,
            ts,
            inter_arrival,
            src_ip,
            dst_ip,
            src_port,
            dst_port,
            protocol,
            ttl,
            ip_header_len,
            tcp_flags,
            udp_length,
            payload_length,
            len(buf)
        ])

print("Rich conversion complete.")
```
